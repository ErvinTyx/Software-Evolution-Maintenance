# Note that Unit_cost, unit_price, profit, cost revenue is string 
it have $ and thousand seperator. to solve this we replace the value using regex

```python
data["Unit_Cost_new"] = data[" Unit_Cost "].str.replace("$", "", regex=False)\
                         .str.replace(",", "", regex=False)\
                         .astype(float)
```

```python
data["Unit_Price_new"] = data[" Unit_Price "].str.replace("$", "", regex=False)\
                         .str.replace(",", "", regex=False)\
                         .astype(float)
```

```python
data["Profit_new"] = data[" Profit "].str.replace("$", "", regex=False)\
                         .str.replace(",", "", regex=False)\
                         .astype(float)
```

```python
data["Cost_new"] = data[" Cost "].str.replace("$", "", regex=False)\
                         .str.replace(",", "", regex=False)\
                         .astype(float)
```

```python
data["Revenue_new"] = data["Revenue"].str.replace("$", "", regex=False)\
                         .str.replace(",", "", regex=False)\
                         .astype(float)
```

formula for profit
profit = (unit cost - unit price) * quantity

formula for revenue
